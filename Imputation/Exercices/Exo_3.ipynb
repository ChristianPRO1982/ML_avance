```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer

# Création d'un dataset avec valeurs manquantes catégoriques
np.random.seed(42)
data = {'gender': np.random.choice(['Male', 'Female', np.nan], size=50),
        'location': np.random.choice(['Urban', 'Suburban', 'Rural', np.nan], size=50),
        'income_level': np.random.choice(['Low', 'Medium', 'High', np.nan], size=50),
        'education_level': np.random.choice(['High School', 'Bachelor', 'Master', np.nan], size=50),
        'target': np.random.randint(0, 2, 50)}

df = pd.DataFrame(data)
df.replace('nan', np.nan, inplace=True)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
# Utiliser SimpleImputer sans spécifier missing_values
imputer = SimpleImputer(strategy='most_frequent')


df = imputer.fit_transform(df)

df = pd.DataFrame(df)
df
```
