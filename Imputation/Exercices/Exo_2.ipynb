```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

np.random.seed(42)
data = {'age': np.random.randint(18, 60, 500),
        'salary': np.random.choice([50000, 60000, np.nan, 80000, 90000], size=500),
        'experience': np.random.randint(1, 15, 500),
        'education': np.random.choice(['High School', 'Bachelor', 'Master', np.nan], size=500),
        'target': np.random.randint(0, 2, 500)}

df = pd.DataFrame(data)
df
```

```python
# Imputation numérique avec SimpleImputer
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("Dataset original :\n", df)
print("\nDataset après imputation :\n", df_imputed)
```
